```python
import pandas as pd

data = {
    "name":["manish",'rahul',"kuldeep","anil"],
    "age":[21, 22, 23 ,24],
    "city":["gujarat", "raniwas", "nagpur", "dausa"]
}

df = pd.DataFrame(data)
print(df)
```

```python
print(df.head())  # First 5 rows
print(df.tail())  # Last 5 rows
print(df.info())  # Summary of the DataFrame
print(df.describe())  # Statistical summary
```

```python
df['country'] = ["india" ,"india","india","india" ]
print(df)
```

```python
df.drop("country" , axis=1 , inplace= True)
print(df)
```

```python
df.rename(columns={"city":"location"} , inplace =True)
print(df)
```

```python
df["age"].fillna(df["age"].mean() , inplace=True)
df['age'] = pd.to_numeric(df['age'], errors='coerce')
print(df)
```

```python
filtered_df = df[df['age'] > 21]
print(filtered_df)
```

```python
import pandas as pd

data = {
    "name":["manish",'rahul',"kuldeep","anil"],
    "age":[21, 22, 23 ,24],
    "city":["gujarat", "raniwas", "nagpur", "dausa"]
}

df = pd.DataFrame(data)
print(df)
```

```python
import seaborn as sns

sns.set(style="whitegrid")

sns.boxplot(x="city", y='age', data=df)
plt.title('Age Distribution by City')
plt.xlabel('City')
plt.ylabel('Age')
plt.show()
```

```python
df["salary"] = [100000 , 50000 , 80000 , 40000] 
print(df)
```

```python
sns.scatterplot(x='age', y='salary', data=df)
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
sns.violinplot(x='city', y='age', data=df)
plt.title('Age Distribution by City')
plt.xlabel('City')
plt.ylabel('Age')
plt.show()
```

```python
# Group by and calculate summary statistics
class_survival = data.groupby('Pclass')['Survived'].mean()
print(class_survival)

# Merge DataFrames (for practice, create a simple DataFrame to merge)
additional_info = pd.DataFrame({
    'Pclass': [1, 2, 3],
    'Description': ['Upper class', 'Middle class', 'Lower class']
})
merged_data = pd.merge(data, additional_info, on='Pclass', how='left')
print(merged_data)
```

```python
import matplotlib.pyplot as plt

# Line plot of passenger ages
plt.figure(figsize=(10, 6))
plt.plot(data['Age'])
plt.title('Passenger Ages')
plt.xlabel('Index')
plt.ylabel('Age')
plt.show()

# Bar plot of survival count
plt.figure(figsize=(10, 6))
data['Survived'].value_counts().plot(kind='bar')
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()

# Histogram of ages
plt.figure(figsize=(10, 6))
data['Age'].plot(kind='hist', bins=20, title='Age Distribution')
plt.xlabel('Age')
plt.show()
```
